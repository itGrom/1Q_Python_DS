```python
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
b = pd.read_csv('creditcard.csv', sep=',')
c = b['Class'].value_counts()
c.plot(kind='bar', logy=True)
plt.show()
```

```python
V1_cl0 = b.loc[b['Class'] == 0,'V1']
V1_cl1 = b.loc[b['Class'] == 1,'V1']

hist_cl0 = plt.hist(V1_cl0, bins=20, color='grey', alpha=0.5, density=True)
hist_cl1 = plt.hist(V1_cl1, bins=20, color='red', alpha=0.5, density=True)

plt.legend(labels=['Class0', 'Class1'])
plt.xlabel('Class')
plt.show()
```
